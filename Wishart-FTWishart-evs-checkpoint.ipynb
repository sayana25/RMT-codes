```python
import numpy as np
import numpy.linalg as nlg
import scipy as sc
from scipy.spatial import cKDTree
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from timeit import default_timer as timer
```

```python
############## Wishart and Fixed Trace WIshart ensemble, larger values of subsystems ####################

n = 10
m = 20 # Matrix size
ens = 50000  # ens
eigvals = []
eigfvals = []
# Generate a single large random GinUE matrix
for i in tqdm(range(ens)):
    A = np.random.normal(0.0, 1.0, (n, m)) + np.random.normal(0.0, 1.0, (n, m))
    A_dag = A.conj().T
    W = np.dot(A,A_dag) #create Wishart matrices
    WF = np.dot(A,A_dag)/np.trace(W) #create fixed trace wishart
    eig = nlg.eigvals(W)
    eigf = nlg.eigvals(WF)
    eigvals.append(eig)
    eigfvals.append(eigf)
```

```python
e1 = np.array(eigvals).flatten()
e2 = np.array(eigfvals).flatten()
```

```python
plt.hist(e1,range = [0,200],  bins=40, density=True, alpha=0.5, color='blue')
```

```python
plt.hist(e2,range = [0,0.4],  bins=40, density=True, alpha=0.5, color='blue')
```

```python
############### In functional form #################################
def LUE_spectrum(n,ens):
    eigvals = []
    # Generate a single large random GOE matrix
    for i in tqdm(range(ens)):
        A = np.random.normal(0.0, 1.0, (n, m)) + np.random.normal(0.0, 1.0, (n, m))
        A_dag = A.conj().T
        W = np.dot(A,A_dag)
        eig = nlg.eigvals(W)
        eigvals.append(eig)
    return eigvals

# Parameters
n = 4 
m = 9 # Matrix size
ens = 50000  # ens

# Compute eigenvalues
eigs =  LUE_spectrum(n,ens)
```

```python
e1 = np.array(eigs).flatten()
```

```python
plt.hist(e1,range = [0,n*m],  bins=40, density=True, alpha=0.5, color='blue')
```
