```python
import numpy as np
from numpy.linalg import eigvals
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
def GinUE(n):
    return np.random.normal(0, 1/np.sqrt(2*n), (n, n)) + 1j * np.random.normal(0, 1/np.sqrt(2*n), (n, n))

def GinOE(n):
    return np.random.normal(0, 1/np.sqrt(2*n), (n, n))
```

```python
# Parameters
n = 60
num_matrices = 5000
ensemble_eigenvalues = []

# Generate Ginibre matrices and compute eigenvalues
for i in tqdm(range(num_matrices)):
    # Compute eigenvalues
    eigenvalues = eigvals(GinUE(n))

    # Sort eigenvalues by absolute value
    sorted_eigenvalues = sorted(eigenvalues, key=np.abs)

    # Extract the middle 40% of the sorted eigenvalues
    middle_start = 0
    middle_end = int(np.ceil(n * 0.4))
    middle_eigenvalues = sorted_eigenvalues[middle_start:middle_end]

    # Append to the list of eigenvalues
    ensemble_eigenvalues.append(middle_eigenvalues)

# Flatten the list of eigenvalues
flat_eigenvalues = np.hstack(ensemble_eigenvalues)

# Plot the eigenvalues
plt.figure(figsize=(6, 6))
plt.scatter(flat_eigenvalues.real, flat_eigenvalues.imag, alpha=0.5)
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Eigenvalues of Ginibre Matrices (40% Middle Segment)')
plt.grid(True)
plt.show()
```
