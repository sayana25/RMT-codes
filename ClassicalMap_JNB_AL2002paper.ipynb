```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
def update_position(X, Y, Z, kappa):
    X_next = Z * np.cos(kappa * X) + Y * np.sin(kappa * X)
    Y_next = Y * np.cos(kappa * X) - Z * np.sin(kappa * X)
    Z_next = -X
    return X_next, Y_next, Z_next

def spherical_coordinates(X, Y, Z):
    theta = np.arccos(Z)
    phi = np.arctan2(Y, X)
    return theta, phi

kappa = 3.0  # Control parameter
steps = 100000  # Number of iterations

# Initial conditions
X, Y, Z = 0.1, 0.1, 0.1  # These will be normalized
X, Y, Z = X / np.linalg.norm([X, Y, Z]), Y / np.linalg.norm([X, Y, Z]), Z / np.linalg.norm([X, Y, Z])
theta, phi = spherical_coordinates(X, Y, Z)

theta_values = [theta]
phi_values = [phi]

for _ in tqdm(range(steps)):
    X, Y, Z = update_position(X, Y, Z, kappa)
    theta, phi = spherical_coordinates(X, Y, Z)
    theta_values.append(theta)
    phi_values.append(phi)

# Plotting theta vs phi
plt.figure(figsize=(8, 6))
plt.scatter(phi_values, theta_values, s=1)
plt.title(r'$\theta$ vs $\phi$ for $\kappa = 3.0$')
plt.xlabel(r'$\phi$')
plt.ylabel(r'$\theta$')
plt.show()
```
