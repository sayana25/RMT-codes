Random Matrix Ensembles

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as nlg
from tqdm.auto import tqdm
```

```python
###############spacing ratios of GOE matrices#####################
def sprat_GOE(n,ens):
    eigvall=[]
    spacing_ratios_all=[]
    for i in tqdm(range(ens)):
        A = np.random.randn(n,n)
        GOE = (A + A.T)/np.sqrt(2*n)
        eig = nlg.eigvals(GOE)
        eigso = np.sort(eig)
        for j in range(n-2):
            sratio = (eigso[j+2]-eigso[j+1])/(eigso[j+1]-eigso[j])
            spacing_ratios_all.append(sratio)
        eigvall.append(eig)
    return eigvall, spacing_ratios_all
    
n = 1000
ens = 50

eigs,sprat = sprat_GOE(n,ens)
```

```python
x = np.linspace(0, 10, 1000)
prgoe = 27*x*(x+1)/(8*(x** 2 + x + 1)**(5/2))  #analytical result for spacing ratios of GOE
prgue = 81*np.sqrt(3)*x**2*(x+1)**2/(4*np.pi*(x ** 2 + x + 1)**(4))  #analytical result for spacing ratios of GUE
prpoi = 1/(x+1)**2  #analytical result for spacing ratios for Poisson(uncorrelated levels)
```

SPACINGS BETWEEN EIGENVALUES

```python
# Plot the spacing ratios
plt.hist(e2,range = [0,8],  bins=80, density=True, alpha=0.3, color='blue')
plt.plot(x,prgoe,alpha=0.7, color='black')
plt.plot(x,prgue,alpha=0.7, color='red')
plt.plot(x,prpoi,alpha=0.7, color='blue')
plt.xlabel('Spacing Ratio')
plt.ylabel('Probability Density')
plt.title('Histogram of Spacing Ratios of Eigenvalues')
plt.grid(True)
plt.show()
```

```python
###############Eigenvalue distribution  of GinUE matrices#####################
def eig_GinUE(n,ens):
    spacing_ratios_alli=[]
    spacing_ratios_allr=[]
    eigvall=[]
    for i in tqdm(range(ens)):
        ginibre_matrix = np.random.normal(0, 1/np.sqrt(2), (n,n)) + 1j * np.random.normal(0, 1/np.sqrt(2), (n,n))
        eig = nlg.eigvals(ginibre_matrix)
        eigsoi = np.sort(np.real(eig))
        eigsor = np.sort(np.imag(eig))
        for j in range(n-2):
    #spacings = np.diff(eigs)
            sratioi = (eigsoi[j+2]-eigsoi[j+1])/(eigsoi[j+1]-eigsoi[j])
            sratior = (eigsor[j+2]-eigsor[j+1])/(eigsor[j+1]-eigsor[j])
            spacing_ratios_alli.append(sratioi)
            spacing_ratios_allr.append(sratior)
        eigvall.append(eig)
    return eigvall,spacing_ratios_alli,spacing_ratios_allr

n = 100
ens = 1000

eigs, sirat, srrat = eig_GinUE(n,ens)
```

```python
e1 = np.array(eigs).flatten()
e2 = np.array(sirat).flatten()
e3 = np.array(srrat).flatten()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(np.real(e1), np.imag(e1), color='blue', alpha=0.7)
plt.title(f"Circular Distribution of Eigenvalues (Ginibre Matrix, size={n}x{n})")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid()
plt.show()
```

```python
def eig_GinOE(n,ens):
    eigvall=[]
###############Eigenvalue distribution  of GinOE matrices#####################
    for i in tqdm(range(ens)):
        GinOE = np.random.normal(0, 1, (n,n)) 
        eig = nlg.eigvals(GinOE)
        eigvall.append(eig)
    return eigvall

n = 100
ens = 500

eigs = eig_GinOE(n,ens)
```

```python
e1 = np.array(eigs).flatten()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(np.real(e1), np.imag(e1), color='grey', alpha=0.3)
plt.title(f"Circular Distribution of Eigenvalues (Ginibre Matrix, size={n}x{n})")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid()
plt.show()
```

```python
############### Eigenvalue distribution  of the Wishart ensemble (slightly different method)#####################
def LOE_eig(n,m,ens):
    eigvall=[]
    for i in tqdm(range(ens)):
        W = np.random.normal(0, 1, (m,n)) 
        WT = W.T
        A = np.matmul(WT, W) #different step instaed of np.dot(), both does matrix multiplication
        eig = nlg.eigvals(A)
        eigvall.append(eig)
    return eigvall
        
m = 20
n = 10
ens = 5000

eigs = LOE_eig(n,m,ens)
```

```python
e1 = np.array(eigs).flatten()
```
