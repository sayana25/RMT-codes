```python
import numpy as np
import numpy.linalg as nlg
import scipy as sc
from scipy.spatial import cKDTree
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from timeit import default_timer as timer
#import seaborn as sns
```

```python
start = timer()
def complex_spacing_ratios(n,ens):
    evt = []
    grdt = []
    for i in tqdm(range(ens)):
        H0 = np.random.normal(0, 1/np.sqrt(2), (n,n))
        # Ginibre matrices with real random gaussian entries with mean = 0, std.dev = 1/sqrt(2)
        #H1 = np.random.normal(0, 1/np.sqrt(2), (n,n)) + 1j * np.random.normal(0, 1/np.sqrt(2), (n,n))
        # Ginibre matrices with real complex gaussian entries with mean = 0, std.dev = 1/sqrt(2)
        eigv = nlg.eigvals(H0)
        evt.extend(eigv)
        grd = []
        kdt_eigv = np.stack([np.real(eigv), np.imag(eigv)], axis=-1)
        kdt = cKDTree(kdt_eigv)
        for j in range(n):
            dists, indices = kdt.query(kdt_eigv[j], 3)
            nns = eigv[indices]
            grd.append((nns[1] - eigv[j]) / (nns[2] - eigv[j]))
        grdt.append(grd)
    return evt,grdt
end = timer()
print(end - start)

n = 300
ens = 1000

eigv,CSR = complex_spacing_ratios(n,ens)
```

```python
def compute_dsff(z_n, tau):
    """Compute the DSFF for a given set of eigenvalues and a complex time tau."""
    tau_star = np.conj(tau)
    sum_exp = np.sum(np.exp(1j * (z_n * tau_star + np.conj(z_n) * tau) / 2)) #equation from ghosh et.al
    return np.abs(sum_exp)**2
```

Crossover between GinOE-GinUE 

```python
start = timer()
def crossover(n,ens,lamb):
    evt = []
    grdt = []
    dsff_values = np.zeros_like(t_values)
    s = 0.3
    for i in tqdm(range(ens)):
        H0 = np.random.normal(0, 1/np.sqrt(2), (n,n))
        H1 = np.random.normal(0, 1/np.sqrt(2), (n,n)) + 1j * np.random.normal(0, 1/np.sqrt(2), (n,n))
        H = 1/(1+lamb**2)*H0 + lamb/(1+lamb**2)*H1
        eigv = nlg.eigvals(H)
        evt.extend(eigv)
        grd = []
        kdt_eigv = np.stack([np.real(eigv), np.imag(eigv)], axis=-1)
        kdt = cKDTree(kdt_eigv)
        for j in range(n):
            dists, indices = kdt.query(kdt_eigv[j], 3)
            nns = eigv[indices]
            grd.append((nns[1] - eigv[j]) / (nns[2] - eigv[j]))
        grdt.append(grd)
        for k, t in enumerate(t_values):
            tau = t + 1j*s 
            dsff_values[k] += compute_dsff(eigv, tau)
    return evt,grdt,dsff_values
end = timer()
print(end - start)

n = 100
ens = 3000
t_values = np.linspace(0.1, 300, 100000)
lamb = 0.005


eigv, CSR, dsff_values =  crossover(n,ens,lamb)
```
